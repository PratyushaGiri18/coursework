```python
#1)
def gcd(a, b):
    while b!= 0:
        a, b = b, a % b
    return a
print(gcd(48, 18))   
```

```python
'''if we use the above code:

If one number is negative, Python’s % still works, but the final result may be negative
#To ensure GCD  always non-negative we can modify like this'''
```

```python
def gcd(a, b):
    if a== 0 and b== 0:
        raise ValueError('gcd undefined for (0, 0)')
    
    while b:
        a, b = b, a % b
    
    return a if a> 0 else -a
gcd(-48, 18)  
```

```python
#2)
#part 1:
def euler_limit(n):
    return (1 + 1/n) ** n
```

```python
#part2:
euler_limit(10**16)
```

```python
# part 3
def euler_sum(n):
    total = 0.0
    factorial = 1  
    for k in range(n):
        if k > 0:
            factorial *= k  
        total += 1 / factorial
        
    return total
euler_sum(4)
```

```python
#part 4
def euler_approx(epsilon):
    total = 0
    factorial = 1
    k = 0

    while True:
        term = 1 / factorial
        
        if term < epsilon:
            break
        
        total += term
        k += 1
        factorial *= k

    return total
```

```python
#part 5
def true_e():
    total = 0
    factorial = 1
    
    for k in range(20):   
        if k > 0:
            factorial *= k
        total += 1 / factorial
    
    return total
# error for euler sum:
def print_euler_sum_error(n):
    e = true_e()
    
    for k in range(1, n+1):
        approx = euler_sum(k)
        error = approx - e
        print(error)
```

```python
# error for euler limit:
def print_euler_lim_error(n):
    e = true_e()
    
    for k in range(1, n+1):
        approx = euler_limit(k)
        error = approx - e
        print(error)
```

```python
print_euler_sum_error(5)
```

```python
#3)
#part 1:
def is_power_of_2(n):
    if n == 1:
        return True
    
    if n % 2 != 0:
        return False
    
    return is_power_of_2(n // 2)
```

```python
#part2
def is_power(b, n):
    if n == 1:
        return True
    
    if n % b != 0:
        return False
    
    return is_power(b, n // b)
```
